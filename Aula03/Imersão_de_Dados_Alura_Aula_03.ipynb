###**Desafio 05:**
Calcular e analisar a correla√ß√£o entre os G e C. (Dica: perceber que est√° sendo plotada uma matriz quadrada e definir como fazer o heatmap disso). Refletir sobre os efeitos biol√≥gicos.


```python
dados_g_c_controle = dados.query('tratamento == "com_controle"').loc[:, 'g-0':'c-99']
```

```python
dados_g_c_controle_corr = dados_g_c_controle.corr()
```

```python
dados_g_c_controle_corr_fatiado = dados_g_c_controle_corr.iloc[0:51 , 772:823]
```

```python
# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(dados_g_c_controle_corr_fatiado, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(dados_g_c_controle_corr_fatiado, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

No grupo de controle, por conta de ser ministrado placebo, a correla√ß√£o entre os nossos G e C √© baixa, assim, as c√©lulas seguem seu comportamento normal, o que n√£o influencia a viabilidade celular de forma direta.

```python
dados_g_c_droga = dados.query('tratamento == "com_droga"').loc[:, 'g-0':'c-99']
```

```python
dados_g_c_droga_corr = dados_g_c_droga.corr()
```

```python
dados_g_c_droga_corr_fatiado = dados_g_c_droga_corr.iloc[0:51, 772:823]
```

```python
# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(dados_g_c_droga_corr_fatiado, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(dados_g_c_droga_corr_fatiado, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

J√° quando as c√©lulas s√£o expostas aos compostos, podemos notar aumento das correla√ß√µes e at√© uma frequ√™ncia no comportamento da viabilidade das c√©lulas, conforme os compostos.

###**Desafio 06:**
Estudar o c√≥digo que plota a matriz de correla√ß√£o (heatmap).

Basicamente, o c√≥digo trabalha em 2 fases, na primeira: segmenta os dados acima da diagonal principal; na segunda: converte todos os valores para 1.
Assim, o heatmap entende onde deve ou n√£o aplicar a m√°scara (somente para valores divergentes de 1).
