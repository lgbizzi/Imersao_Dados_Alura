```python
import numpy as np

corr = dados.loc[:, 'g-0':'g-50'].corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

###**Desafio 05:**
Calcular e analisar a correlação entre os G e C. (Dica: perceber que está sendo plotada uma matriz quadrada e definir como fazer o heatmap disso). Refletir sobre os efeitos biológicos.


```python
dados_g_c_controle = dados.query('tratamento == "com_controle"').loc[:, 'g-0':'c-99']
```

```python
dados_g_c_controle_corr = dados_g_c_controle.corr()
```

```python
dados_g_c_controle_corr_fatiado = dados_g_c_controle_corr.iloc[0:51 , 772:823]
```

```python
# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(dados_g_c_controle_corr_fatiado, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(dados_g_c_controle_corr_fatiado, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

No grupo de controle, por conta de ser ministrado placebo, a correlação entre os nossos G e C é baixa, assim, as células seguem seu comportamento normal, o que não influencia a viabilidade celular de forma direta.

```python
dados_g_c_droga = dados.query('tratamento == "com_droga"').loc[:, 'g-0':'c-99']
```

```python
dados_g_c_droga_corr = dados_g_c_droga.corr()
```

```python
dados_g_c_droga_corr_fatiado = dados_g_c_droga_corr.iloc[0:51, 772:823]
```

```python
# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(dados_g_c_droga_corr_fatiado, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(dados_g_c_droga_corr_fatiado, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

Já quando as células são expostas aos compostos, podemos notar aumento das correlações e até uma frequência no comportamento da viabilidade das células, conforme os compostos.

###**Desafio 06:**
Estudar o código que plota a matriz de correlação (heatmap).

Basicamente, o código trabalha em 2 fases, na primeira: segmenta os dados acima da diagonal principal; na segunda: converte todos os valores para 1.
Assim, o heatmap entende onde deve ou não aplicar a máscara (somente para valores divergentes de 1).
